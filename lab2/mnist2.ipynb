# Importing Libraries and Downloading mnist Data

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
(x_train, y_train), (x_test, y_test)=tf.keras.datasets.mnist.load_data()
```

**Splitting Training Data into validation and Training**

```python
x_validate,y_validate=x_train[55000:60000],y_train[55000:60000]
x_train,y_train=x_train[0:55000],y_train[0:55000]
```

**Running DNN using 32 neurons in hidden layer and using activation function sigmoid**

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=[28,28]),
    keras.layers.Dense(32,activation='sigmoid'),
    keras.layers.Dense(10,activation='softmax')
])
sgd=keras.optimizers.SGD(lr=0.1)
model.compile(optimizer='sgd',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(x_train,y_train,validation_data=(x_validate,y_validate),epochs=20)

test_loss, test_acc = model.evaluate(x_test,  y_test, verbose=2)

print('\nTest accuracy:', test_acc)
```

# Running After Normalising the Data

```python
nx_train=x_train/255
nx_test=x_test/255
nx_validate=x_validate/255
nmodel=keras.Sequential([
    keras.layers.Flatten(input_shape=[28,28]),
    keras.layers.Dense(32,activation='sigmoid'),
    keras.layers.Dense(10,activation='softmax')
])
sgd=keras.optimizers.SGD(lr=0.1)
nmodel.compile(optimizer='sgd',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
nmodel.fit(nx_train,y_train,epochs=20,validation_data=(nx_validate,y_validate))

ntest_loss, ntest_acc = nmodel.evaluate(nx_test,  y_test, verbose=2)

print('\nTest accuracy:', ntest_acc)
```

**Using Relu as activation function**

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=[28,28]),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dense(10,activation='softmax')
])
sgd=keras.optimizers.SGD(lr=0.001)
model.compile(optimizer='sgd',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(nx_train,y_train,epochs=50,validation_data=(nx_validate,y_validate))

test_loss, test_acc = model.evaluate(nx_test,  y_test, verbose=2)

print('\nTest accuracy:', test_acc)
```

**Using Leaky Relu as activation function**

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=[28,28]),
    keras.layers.Dense(64,activation=keras.layers.LeakyReLU(alpha=0.1)),
    keras.layers.Dense(64,activation=keras.layers.LeakyReLU(alpha=0.1)),
    keras.layers.Dense(64,activation=keras.layers.LeakyReLU(alpha=0.1)),
    keras.layers.Dense(10,activation='softmax')
])
sgd=keras.optimizers.SGD(lr=0.001)
model.compile(optimizer='sgd',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(nx_train,y_train,epochs=50,validation_data=(nx_validate,y_validate))

test_loss, test_acc = model.evaluate(nx_test,  y_test, verbose=2)

print('\nTest accuracy:', test_acc)
```

**Leaky Relu is better activation function as accuracy of model is 97.71% which is better than relu and tanh activation functions**
