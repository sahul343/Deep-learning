# Fetching the Data

```python
from nltk.corpus import stopwords
from sklearn.datasets import fetch_20newsgroups
from nltk.tokenize import word_tokenize,RegexpTokenizer
import gensim
from gensim.models import Word2Vec
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense,Dropout,Activation
from sklearn.preprocessing import StandardScaler
from keras.layers import LeakyReLU
corpus=fetch_20newsgroups(subset='all')
stop_words=set(stopwords.words('english'))
```

# Pre-processing the Data

```python
processed_data=[]
tokenizer=RegexpTokenizer(r'\w+')
for para in corpus.data:
  tokens=tokenizer.tokenize(para.lower())
  processed_para=[word for word in tokens if word not in stop_words and word.isalpha()]
  processed_data.append(processed_para)
```

# Converting the words to vectors using word2vec

```python
skip_gram=Word2Vec(processed_data,size=100,sg=1,window=5,min_count=10)
CBOW=Word2Vec(processed_data,size=100,sg=0,window=5,min_count=10)
```

**Vocabulary Size** : 23637

```python
#skip_gram model
print("vocabulary size: ",len(skip_gram.wv.vocab))
#cbow model
print("vocabulary size: ",len(CBOW.wv.vocab))
```

# TSNE plot for top 20 words similar to the words given in question

```python
words=["baseball" ,"software","police", "government", "circuit", "car"]
colors=['r','b','c','g','m','y']
similarsg=[]
similarcbow=[]
sgwords=[]
cbowwords=[]
for word in words:
  sw1=skip_gram.most_similar(word,topn=20)
  sw2=CBOW.most_similar(word,topn=20)
  s1=[skip_gram.wv[x[0]] for x in sw1]
  s2=[CBOW.wv[x[0]] for x in sw2]
  similarsg+=s1
  similarcbow+=s2
  sgwords+=[x[0] for x in sw1]
  cbowwords+=[x[0] for x in sw1]

sg_2d=TSNE(n_components=2).fit_transform(similarsg)
cbow_2d=TSNE(n_components=2).fit_transform(similarcbow)
```

# Representing Each document as datapoint


#for skip gram model

```python
Datapoint_sgmodel=[]
for document in processed_data:
  datapoint=np.zeros((100,))
  count=0
  for word in document:
    if word in skip_gram.wv.vocab:
      datapoint+=skip_gram.wv[word]
      count+=1
  if(count!=0):
    datapoint/=count;
    Datapoint_sgmodel.append(datapoint)
```

#for cbow model

```python
Datapoint_cbowmodel=[]
for document in processed_data:
  datapoint=np.zeros((100,))
  count=0
  for word in document:
    if word in CBOW.wv.vocab:
      datapoint+=CBOW.wv[word]
      count+=1
  if(count!=0):
    datapoint/=count;
    Datapoint_cbowmodel.append(datapoint)
```

```python
Datapoint_cbowmodel=np.array(Datapoint_cbowmodel)
Datapoint_sgmodel=np.array(Datapoint_sgmodel)
```

# Part-B of question 

# skip Gram

```python
class dnn:
  def __init__(self,sizes):
    self.num_layers=len(sizes)
    self.sizes=sizes
    self.model=keras.Sequential()
    self.model.add(Dense(sizes[0],activation=LeakyReLU(alpha=0.1),input_dim=100))
    for i in range(1,len(sizes)-1):
      self.model.add(Dense(sizes[i],activation=LeakyReLU(alpha=0.1)))
    self.model.add(Dense(sizes[-1],activation='softmax'))
    self.model.compile( loss='sparse_categorical_crossentropy',
                        optimizer='adam',
                        metrics=['accuracy']
    )
  def train(self,x_train,y_train,x_val,y_val):
    self.Metadata=self.model.fit(x_train,y_train,epochs=50,
                   validation_data=(x_val,y_val),
                   shuffle=True,
                   batch_size=20
                  )
  def predict(self,x_test,y_test):
    test_loss, test_acc = self.model.evaluate(x_test,  y_test, verbose=2)
    return test_acc
    
```

```python
dnnskip_gram=dnn([100,40,40,20])
```

```python
Datasize=len(Datapoint_sgmodel)
Y=corpus.target
x_train=Datapoint_sgmodel[0:int(0.7*Datasize)]
x_val=Datapoint_sgmodel[int(0.7*Datasize):int(0.8*Datasize)]
x_test=Datapoint_sgmodel[int(0.8*Datasize):Datasize]
y_train=Y[0:int(0.7*Datasize)]
y_val=Y[int(0.7*Datasize):int(0.8*Datasize)]
y_test=Y[int(0.8*Datasize):Datasize]
dnnskip_gram.train(x_train,y_train,x_val,y_val)
print("Skip-Gram Model accuracy=", dnnskip_gram.predict(x_test,y_test))
```

# Plottings For skip-gram Model

```python
plt.ylabel('classification accuracy ')
plt.xlabel('Iterations')
plt.plot(dnnskip_gram.Metadata.history['acc'],'g')
plt.show()

plt.ylabel('Classification Error ')
plt.xlabel('Iterations')
plt.plot(1-np.array(dnnskip_gram.Metadata.history['acc']),'r')
plt.show()


plt.ylabel('Loss')
plt.xlabel('Iterations')
plt.plot(dnnskip_gram.Metadata.history['loss'],'r')
plt.show()
```

```python

print('training data:\n')
print("classification accuracy:",dnnskip_gram.Metadata.history['acc'][-1])

print('testing data:\n')
print('clssification accuracy:',dnnskip_gram.predict(x_test,y_test))
```
